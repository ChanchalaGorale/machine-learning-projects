```python
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
    database="test_db",
    port=3306
)

cursor = conn.cursor()

cursor.execute("""
CREATE TABLE employees (
    id INT PRIMARY KEY,
    name VARCHAR(100),
    department VARCHAR(100),
    position VARCHAR(100),
    salary FLOAT
)
""")

conn.commit()
conn.close()
```

```python
def connect_db():
  return mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
    database="test_db",
    port=3306
)
```

```python
def check_employee(id):
  conn = connect_db()

  sql = 'SELECT * FROM employees WHERE id=%s'

  cursor = conn.cursor(buffered=True)

  data = (id, )

  cursor.execute(sql, data)

  employee = cursor.fetchone()

  cursor.close()

  return employee is not None
```

```python
def add_employee():
  conn = connect_db()

  Id = input("Enter employee id: ")

  if check_employee(Id):
    print("Warning: Employee already exists.")

    return

  else:
    Name = input("Enter employee name: ")
    Post = input("Enter employee post: ")
    Salary = input("Enter employee salary: ")


    sql = 'INSERT INTO employees (id, name, position, salary) VALUES (%s, %s, %s, %s)'

    data = (Id, Name, Post, Salary)

    cursor = conn.cursor()

    try:
      cursor.execute(sql, data)

      conn.commit()

      print("New employee created!")
    except mysql.connector.Error as err:
      print(f'Error occured: {err}')

      conn.rollback()

    finally:
      cursor.close()
```

```python
def remove_employee():
  conn = connect_db()

  Id = input("Enter employee id: ")

  if not check_employee(Id):
    print("Employee does not exist.")

    return
  else:
    sql = 'DELETE FROM employees WHERE id=%s'

    data = (Id,)

    cursor = conn.cursor()

    try:
      cursor.execute(sql, data)

      conn.commit()

      print("Emploee deleted!")
    except mysql.connector.Error as err:
      print(f'Error occured: {err}')

      conn.rollback()

    finally:
      cursor.close()
```

```python
def promote_employee():
  conn = connect_db()

  Id= input("Employee Id: ")

  if not check_employee(Id):
    print('Employee does not exist.')

    return
  else:
    try:
      Amount = input("Enter increas in salary: ")

      sql_select = 'SELECT salary FROM employees WHERE id=%s'

      data = (Id,)

      cursor = conn.cursor()

      cursor.execute(sql_select, data)

      current_salary = cursor.fetchone()[0]

      new_salary = current_salary + Amount

      sql_update = 'UPDATE employees SET salary=%s WHERE id=%s'

      data_update = (new_salary, Id)

      cursor.execute(sql_update, data_update)

      conn.commit()

      print("Employee promoted!")

    except (ValueError, mysql.connector.Error) as err:
      print(f'Error: {err}')

    finally:
      cursor.close()
```

```python
def display_employee():
  conn = connect_db()

  try:
    sql = 'SELECT * FROM employees'

    cursor = conn.cursor()

    cursor.execute(sql)

    employees = cursor.fetchall()

    for emp in employees:
      print("Emploee Id: ", emp[0])
      print("Emploee Name: ", emp[1])
      print("Emploee Post: ", emp[2])
      print("Emploee Salary: ", emp[3])
      print()
      print("__________________________")
      print()
  except mysql.connector.Error as err:
    print(f'Error: {err}')
  finally:
    cursor.close()
```

```python
def menu():
  while True:
    print("\nWelcome to Employee Management Record")
    print("Press:")
    print("1 to Add Employee")
    print("2 to Remove Employee")
    print("3 to Promote Employee")
    print("4 to Display Employees")
    print("5 to Exit")

    ch = input("Enter menu: ")

    if ch == '1':
      add_employee()
    elif ch == '2':
      remove_employee()
    elif ch == '3':
      promote_employee()
    elif ch == '4':
      display_employee()
    elif ch == '5':
      print("Exiting the program.")
      break
    else:
      print("Invalid choice!")
```
