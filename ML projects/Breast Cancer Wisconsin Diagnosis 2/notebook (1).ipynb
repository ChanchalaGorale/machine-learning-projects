# Breast Cancer Wisconsin Diagnosis

```python
import os
import kagglehub

path = kagglehub.dataset_download("uciml/breast-cancer-wisconsin-data")

print("Path to dataset files:", path)
```

```python
os.listdir(path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv(os.path.join(path, 'data.csv'))

df.head()
```

```python
df.info()
```

```python
df.drop(['Unnamed: 32', 'id'], axis=1)

df.head()
```

```python
def diagnosis_value(diagnosis):
  if diagnosis == "M":
    return 1
  else:
    return 0
df['diagnosis'] = df["diagnosis"].apply(diagnosis_value)
```

```python
sns.lmplot(x="radius_mean", y="texture_mean", data=df, hue="diagnosis")
```

```python
sns.lmplot(x="smoothness_mean", y="compactness_mean", data=df, hue="diagnosis")
```

```python
X = df.drop(['diagnosis', "id", "Unnamed: 32"], axis =1)
y = df['diagnosis']
```

```python
X.isnull().any()
```

```python
y.isnull().any()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=11) # updated from 13 to 11

knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test) # score improved 0.10 %
```

```python
neighbors = []
cv_scores = []

from sklearn.model_selection import cross_val_score

for k in range(1, 51, 2):
    neighbors.append(k)
    knn = KNeighborsClassifier(n_neighbors = k)
    scores = cross_val_score(
        knn, X_train, y_train, cv = 10, scoring = 'accuracy')
    cv_scores.append(scores.mean())
```

```python
MSE = [1-x for x in cv_scores]

optimal_k = neighbors[MSE.index(min(MSE))]
print('The optimal number of neighbors is % d ' % optimal_k)

plt.figure(figsize = (10, 6))
plt.plot(neighbors, MSE)
plt.xlabel('Number of neighbors')
plt.ylabel('Misclassification Error')
plt.show()
```
