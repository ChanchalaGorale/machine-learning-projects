# Heart Disease Prediction

```python
import pandas as pd

df = pd.read_csv("/content/framingham.csv")

df.head()
```

```python
df.info()
```

```python
df= df.drop("education", axis=1)
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
target = "TenYearCHD"
```

```python
df[target].value_counts()
```

```python
df.corr()[target].drop(target)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = df.drop([target], axis=1)
y=df[target]

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)

st_scl = StandardScaler()
x_train = st_scl.fit_transform(x_train)
x_test = st_scl.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(x_train, y_train)
```

```python
coef_df = pd.DataFrame({
    "coef": model.coef_[0], "feature": X.columns
})

coef_df
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import recall_score

recall_score(y_test, y_pred)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import recall_score
from collections import Counter

counter = Counter(y_train)

model =XGBClassifier(scale_pos_weight=counter[0] / counter[1], random_state=13)

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print(recall_score(y_test, y_pred))
```
