```python
from prophet import Prophet
```

```python
def sales_predict(id, df, time):
  df = df[df["Store"]==id]
  df = df[["Date", "Sales"]].rename(columns={"Date": "ds", "Sales": "y"})
  df=df.sort_values("ds")

  model = Prophet()
  model.fit(df)
  future = model.make_future_dataframe(periods=time)
  forecast = model.predict(future)

  model.plot(forecast, xlabel="Date", ylabel="Sales")
  model.plot_components(forecast)
```

```python
sales_predict(10, df_all, 60)
```

```python
def sales_predict_with_holidays(id, df, holiday, time):
  df = df[df["Store"]==id]
  df = df[["Date", "Sales"]].rename(columns={"Date": "ds", "Sales": "y"})
  df=df.sort_values("ds")

  model = Prophet(holidays=holiday)
  model.fit(df)
  future = model.make_future_dataframe(periods=time)
  forecast = model.predict(future)

  model.plot(forecast, xlabel="Date", ylabel="Sales")
  model.plot_components(forecast)
```

```python
df_school_holidays = df_all[df_all["SchoolHoliday"]==1].loc[:, "Date"].values
df_state_holidays = df_all[(df_all["StateHoliday"]=="a") |(df_all["StateHoliday"]=="b")|(df_all["StateHoliday"]=="c")].loc[:, "Date"].values
```

```python
df_school_holidays.shape, df_state_holidays.shape
```

```python
df_state_holidays = pd.DataFrame({"ds":pd.to_datetime(df_state_holidays), "holiday":"state"})

df_school_holidays = pd.DataFrame({"ds":pd.to_datetime(df_school_holidays), "holiday":"school"})
```

```python
df_holiday = pd.concat([df_state_holidays, df_school_holidays])
df_holiday.head()
```

```python
sales_predict_with_holidays(6, df_all, df_holiday, 90)
```
