# Find K for Kmeans model

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
scaled_data.shape
```

```python
scaled_data
```

```python
score_1 = []

range_values= range(1, 20)

for i in range_values:
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_data)
    score_1.append(kmeans.inertia_)
```

```python
score_1
```

```python
plt.plot(score_1, "bx-")
plt.xlabel("Values of K")
plt.ylabel("Inertia")
```

We will proceed with 7 clusters

# Tensorflow model

```python
from tensorflow.keras.layers import Dense, Input, Add, Activation, ZeroPadding2D, BatchNormalization, Flatten, Conv2D, AveragePooling2D, MaxPooling2D, Dropout
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.initializers import glorot_uniform
from tensorflow.keras.optimizers import SGD
```

```python
input_df= Input(shape=(17,))
```

```python
x= Dense(7, activation="relu")(input_df)
```

```python
x= Dense(500, activation="relu", kernel_initializer="glorot_uniform")(x)
```

```python
x= Dense(2000, activation="relu", kernel_initializer="glorot_uniform")(x)
```

```python
encoder = Dense(10, activation="relu",  kernel_initializer="glorot_uniform")(x)
```

```python
x= Dense(500, activation="relu", kernel_initializer="glorot_uniform")(encoder)
```

```python
decoder = Dense(17, activation="relu", kernel_initializer="glorot_uniform")(x)
```

```python
autoencoder = Model(inputs=input_df, outputs=decoder)
```

```python
encoder_model = Model(inputs=input_df, outputs=encoder)
```

```python
autoencoder.compile(optimizer="adam", loss="mse")
```

```python
autoencoder.fit(scaled_data, scaled_data, epochs=100, batch_size=32, verbose=True)
```

```python
autoencoder.summary()
```
