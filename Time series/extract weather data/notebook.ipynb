# Extract Weather Data from Google

```python
import requests
from bs4 import BeautifulSoup

city = "lucknow"

url = "https://www.google.com/search?q="+"weather"+city

html = requests.get(url).content

soup = BeautifulSoup(html, 'html.parser')

print(soup.prettify())

temp_div = soup.find('div', attrs={'class': 'BNeawe iBp4i AP7Wnd'})

if temp_div:
    temp = temp_div.get_text()

    data = temp.split('\n')
    time = data[0]
    sky = data[1]

    listdiv = soup.findAll('div', attrs={'class': 'BNeawe s3v9rd AP7Wnd'})

    if listdiv:
      strd = listdiv[5].get_text()

      pos = strd.find('Wind')
      other_data = strd[pos:]

      print("Temperature is", temp)
      print("Time: ", time)
      print("Sky Description: ", sky)
      print(other_data)
    else:
      print("Div list not found.")
else:
    print("Temperature div not found.")
```
